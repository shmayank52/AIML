```python
import pandas as pd
from google.colab import files
uploaded=files.upload()
```

```python
import io
```

```python
df=pd.read_csv(io.BytesIO(uploaded['Drishय (1).csv']))
```

```python
# Remove rows with missing values
df.dropna(axis=0, inplace=True)
```

```python
# Remove columns with missing values
df.dropna(axis=1, inplace=True)
```

```python
# Fill missing numerical data by mean
df['Strength'].fillna(df['Strength'].mean(), inplace=True)
```

```python
# Fill missing numerical data by median
df['Strength'].fillna(df['Strength'].median(), inplace=True)
```

```python
# Fill missing categorical data by mode
df['Strength'].fillna(df['Strength'].mode(), inplace=True)
```

```python
df.infer_objects(copy=False)
```

```python
# Forward fill
df.ffill(inplace=True)

# Backward fill
df.bfill(inplace=True)
```

```python
from scipy import stats
import numpy as np

# Z-score
z_scores = np.abs(stats.zscore(df['Strength']))
df = df[(z_scores < 3)]  # Keep rows with Z-score less than 3
```

```python
Q1 = df['Strength'].quantile(0.25)
Q3 = df['Strength'].quantile(0.75)
IQR = Q3 - Q1

# Filter out outliers
df = df[~((df['Strength'] < (Q1 - 1.5 * IQR)) | (df['Strength'] > (Q3 + 1.5 * IQR)))]
```
